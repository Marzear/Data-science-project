```python
# Run in jupyter notebook
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('dick data/training_POS.csv')
def scatter_plot(x,y):
    ax = data.query('label==1 & POS != "P" & POS != "C"').plot.scatter(x=x,y=y,color='#ef5350')
    data.query('label==0 & POS != "P" & POS != "C"').sample(500).plot.scatter(x=x,y=y,color='#26C6DA',ax=ax,figsize=(13,7))
```

```python
scatter_plot('OBP','SLG')
```
